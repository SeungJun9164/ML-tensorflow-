```python
# matplotlib : pyplot 객체를 사용하여 데이터를 표시
import matplotlib.pyplot as plt
X = range(100)
Y = [value**2 for value in X]
plt.plot(X, Y)
plt.show()
```

```python
import numpy as np 
x_1 = range(100)
y_1 = [np.cos(value) for value in X]

x_2 = range(100)
y_2 = [np.sin(value) for value in X]

plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.show()
```

```python
fig = plt.figure() # figure 반환
fig.set_size_inches(10, 5)# 크기 지정
ax_1 = fig.add_subplot(1,2,1) # 두개의 plot생성
ax_2 = fig.add_subplot(1,2,2)

ax_1.plot(x_1, y_1, c="b")
ax_2.plot(x_2, y_2, c="g")
plt.show()
```

```python
# set color
plt.plot(x_1, y_1, color="#eeefff")
plt.plot(x_2, y_2, color="r") # r : red / b : blue / g : green

plt.show()
```

```python
# set linestyle : ls || linestyle로 사용
plt.plot(x_1, y_1, c="b", linestyle="dashed") 
plt.plot(x_2, y_2, c="r", ls="dotted")

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X]

X_2 = range(100)
Y_2 = [value + 100 for value in X]
```

```python
# title
plt.plot(X_1, Y_1, color="b", linestyle="dashed")
plt.plot(X_2, Y_2, color="r", linestyle="dotted")

plt.title("Two lines")
plt.show()
```

```python
# set legend : 범례 표시
plt.plot(X_1, Y_1, color="b", linestyle="dashed", label='line_1')
plt.plot(X_2, Y_2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=True, loc="lower right")
```

```python
# scatter : 산포도
data_1 = np.random.rand(512, 2)
data_2 = np.random.rand(512, 2)
```

```python
plt.scatter(data_1[:,0], data_1[:,1], c="b", marker="x") # marker : 모양 표시
plt.scatter(data_2[:,0], data_2[:,1], c="r", marker="o")

plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2 
plt.scatter(x, y, s=area, c=colors, alpha=0.5) # s : 크기 지정
plt.show()
```

```python
# bar chart
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]] # y의 값이 됨

X = np.arange(0,8,2)
```

```python
# x는 0 1 2...이런식으로 가는데 거기에 0.00 를 더해서 나타냄
plt.bar(X + 0.00, data[0], color = 'b', width = 0.50) 
plt.bar(X + 0.50, data[1], color = 'g', width = 0.50)
plt.bar(X + 1.0, data[2], color = 'r', width = 0.50)
plt.xticks(X+0.50, ("A","B","C", "D"))
plt.show()
```

```python
# histogram
X = np.random.randn(1000)
plt.hist(X, bins=100) # bins :데이터를 나누는 갯수
plt.show()
```

```python
# boxplot
data = np.random.randn(100, 5)
plt.boxplot(data)
plt.show()
```
