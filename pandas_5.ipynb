```python
import pandas as pd
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
         'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
         'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
         'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
         'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
# Groupby : SQL groupby 명령어와 같음
# split -> apply -> combine 과정을 거쳐 연산
# groupby(기준)[적용받는 열].apply
df.groupby("Team")["Points"].sum()
```

```python
# 한 개 이상의 열을 묶을 수 있다.
df.groupby(["Team", "Year"])["Points"].sum()
```

```python
# Hierarchical index : 두 개의 열로 groupby를 할 경우, index가 두개 생성
h_index = df.groupby(["Team", "Year"])["Points"].sum()
h_index
```

```python
h_index.index
```

```python
h_index["Devils":"Kings"]
```

```python
# unstack : group으로 묶여진 데이터를 matrix 형태로 전환
h_index.unstack()
```

```python
h_index
```

```python
# swaplevel : index level을 변경
h_index.swaplevel()
```

```python
# index level을 기준으로 기본 연산 수행 가능
h_index.sum(level=0)
```

```python
h_index.sum(level=1)
```

```python
# grouped : groupby에 의해 splic된 상태를 추출
# split - apply - combine 형태에서 split상태를 그대로 저장
grouped = df.groupby("Team")
for name, group in grouped: # name : devlis, kings, ... / group : dataFrame
    print(name)
    print(group)
```

```python
# filter : 특정 조건으로 데이터를 검색할 때 사용
# filter안에는 boolean 조건이 존재해야함
df.groupby('Team').filter(lambda x: len(x) >= 3)
```

```python
df_phone = pd.read_csv("https://www.shanelynn.ie/wp-content/uploads/2015/06/phone_data.csv")
```

```python
df_phone.head()
```

```python
# import dateutil : date타입을 string이 아닌 date타입으로 변경
import dateutil
df_phone['date'] = df_phone['date'].apply(dateutil.parser.parse, dayfirst=True)
df_phone.head()
```

```python
df_phone.groupby('month')['duration'].sum()
```

```python
df_phone[df_phone['item'] == 'call'].groupby('month')['duration'].sum()
# item 이 call인 것만 불러와 month 기준으로 묶어라
```

```python
df_phone.groupby(['month', 'item'])['duration'].sum()
```

```python
df_phone.groupby(['month', 'item'])['date'].count().unstack()
```

```python
# agg : 여러개의 함수 한번에 적용하기 위해
df_phone.groupby(['month', 'item']).agg({'duration':sum,      
                                     'network_type': "count", 
                                     'date': 'first'})    
```

```python
df_phone.groupby(['month', 'item']).agg({'duration': [min],      
                                     'network_type': "count", 
                                     'date': [min, 'first', 'nunique']})
```
